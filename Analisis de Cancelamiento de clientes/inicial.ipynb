```python


# clientes de contraro mensual canselamientos
# ofreser decuento en los planes anuais o trimestrais
tabela = tabela[tabela["duracao_contrato"]
                != "Monthly"]  # esto es una condicion
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))

# clientes  que ligan mas de 4 veces para el coocenter cancela
# ofreser una estrategia cn el cliente en 3 ligacoes
tabela = tabela[tabela["ligacoes_callcenter"]
                <= 4]
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))

# clientes quesn se atrasaram mais de 20 dias cancelan
# politica de resolver atrazons en hasta 10 dias
tabela = tabela[tabela["dias_atraso"]
                <= 20]
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```
