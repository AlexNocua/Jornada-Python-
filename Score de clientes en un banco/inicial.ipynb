### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

```python
# paso a paso
# Passo 0 - entender el problema de la empresa
# paso 1 - importar as bases de dados
# paso 2 preparar la base de datos para ia
# paso 3 criar um modelo de ia -> score de credito: ruim , bom, medio
# paso 4 escolher o melhor modelo
# paso 5 - usar a nossa ia para fazer novas previsoes
```

```python
# paso 1 - importar as bases de dados
import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
display(tabela.info())
```

```python

# mix_credito
# score_credito
# tratamiento de esas 3 colunas porque son textos, esto tiene que ser trasformado en numeros para hacer porcesos de IA
# de toda la biblioteca de la seccion preprocessing cojo un libro.
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()  # codificador aplica en la columna tatata

# Profissao
# no se pueden hacer las cpdoficaciones al mismo tiempo debido que a cada coluna necesita una codificacion diferente
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador.fit_transform(
    tabela["score_credito"])
display(tabela)
```

```python
# aprendizado de maquina
# y columna que eu quero prever
y = tabela["score_credito"]
# columnas que eu quero para fazer a previcao
tabela = tabela.drop(columns=["score_credito", "id_cliente"])
x = tabela
```

```python
display(tabela)
display(x)
display(y)
```

```python
from sklearn.model_selection import train_test_split

# el esta escogiendo datos al azar de la base de datos y las divide en cada una de esas variables
# test_size=0.3 esto se defien mediante el tts con el find de colocar 3 porciento de tes y 70 porciento de entreno
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
# criar a IA
# arvore de decicao Randomforest
# KNN neirest neigvors vecinos proximos Kneigbors

# importta

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
# crea a ia#

modelo_arvoredecissao = RandomForestClassifier()
modelo_neighbor = KNeighborsClassifier()


# entrena a ia

modelo_arvoredecissao.fit(x_treino, y_treino)
modelo_neighbor.fit(x_treino, y_treino)
# treinar os modelos
```

```python
# testar os modelos
from sklearn.metrics import accuracy_score
# con esto se cambian los formatos para que funcionen
previssao_arvore = modelo_arvoredecissao.predict(x_teste)
presissao_knn = modelo_neighbor.predict(x_teste.to_numpy())


# acurancia

print(accuracy_score(y_teste, previssao_arvore))
print(accuracy_score(y_teste, presissao_knn))
```

```python
# fazer novas previcoes

# importar novos clients
# codificar os novos clientes

tabela_novos_clientes = pd.read_csv("novos_clientes.csv")
print(tabela_novos_clientes)
tabela_novos_clientes["profissao"] = codificador.fit_transform(tabela_novos_clientes["profissao"])
tabela_novos_clientes["mix_credito"] = codificador.transform(tabela_novos_clientes["mix_credito"])


display(tabela_novos_clientes)

previsoes = modelo_arvoredecissao.predict(tabela_novos_clientes.to_numpy())
print(previsoes)
```
